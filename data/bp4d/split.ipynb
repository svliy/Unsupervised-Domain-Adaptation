```python
import os
import pandas as pd
import numpy as np
```

```python
validation_set = pd.DataFrame()
df = pd.read_csv('./total.csv')
df
```

```python
unique_persons = df['person'].drop_duplicates()
```

```python
len(unique_persons)
```

```python
for person in unique_persons:
    person_df = df[df['person'] == person]  # 筛选出当前 person 的所有样本
    
    # 计算需要抽取的样本数量
    num_samples = int(len(person_df) * 0.1)
    
    # 如果样本数量为 0，跳过当前 person
    if num_samples == 0:
        print(f"警告：{person} 的样本数量太少，无法抽取 10%，跳过。")
        continue
    
    sampled_df = person_df.sample(n=num_samples, replace=False)  # 抽取指定数量的样本
    
    validation_set = pd.concat([validation_set, sampled_df], ignore_index=True)  # 将抽取的样本添加到验证集中
```

```python
validation_set
```

```python
validation_set.to_csv('total_val.csv', index=None)
```

```python
len(validation_set['person'].drop_duplicates())
```

```python
# 获取验证集中的 img-path 列表
img_paths_to_remove = validation_set['img_path'].tolist()

# 从原始 DataFrame df 中删除行
df = df[~df['img_path'].isin(img_paths_to_remove)]

# 打印移除行后的 DataFrame df 的大小，以进行验证
print("移除行后的 DataFrame df 的大小：", df.shape)
```

```python
df.to_csv('total_train.csv', index=None)
```
